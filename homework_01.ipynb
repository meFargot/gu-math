Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20). Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
def vec_len(vec):
    """ Расчитывает длину вектора через цикл
    """
    dims = len(vec)
    sum_sq = 0
    for i in range(dims):
        sum_sq += vec[i] ** 2
    return np.sqrt(sum_sq)
```

```python
def vec_len_2(vec):
    """ Расчитывает длину вектора без цикла
    """
    return np.sqrt(np.sum([coord ** 2 for coord in vec]))
```

##### Проверка

```python
# зададим вектора
x1 = np.array([20,20,20])
x2 = np.array([0,0,-20])
```

```python
# расчет по формуле
np.sqrt(x1[0]**2 + x1[1]**2 + x1[2]**2)
```

```python
# расчет с помощью написанных функций
vec_len(x1)
```

```python
vec_len_2(x1)
```

Напишите код на Python, реализующий построение графиков:
1) окружности,
2) эллипса,
3) гиперболы.

##### Окружность

```python
x = []
y1 = []
y2 = []
R = 3
for i in range(-5000, 5001):
    x_ = i/1000
    x.append(x_)
    y1.append(np.sqrt(R**2 - x_**2))
    y2.append(-np.sqrt(R**2 - x_**2))
plt.figure(figsize=(5,5))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.plot(x,y1, color='blue')
plt.plot(x,y2, color='blue')
plt.xlabel('x')
plt.ylabel('y')
```

##### Эллипс

```python
x = []
y1 = []
y2 = []
a = 2
b = 1
for i in range(-5000, 5001):
    x_ = i/1000
    x.append(x_)
    y1.append(np.sqrt( b*(1 - (x_**2/a**2)) ))
    y2.append(-np.sqrt( b*(1 - (x_**2/a**2)) ))
plt.figure(figsize=(5,5))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.plot(x,y1, color='blue')
plt.plot(x,y2, color='blue')
plt.xlabel('x')
plt.ylabel('y')
```

##### Гипербола

```python
x = []
y1 = []
y2 = []
a = 1
b = 1
for i in range(-5000, 5001):
    x_ = i/1000
    x.append(x_)
    y1.append(np.sqrt( b*((x_**2/a**2) - 1) ))
    y2.append(-np.sqrt( b*((x_**2/a**2) - 1) ))
plt.figure(figsize=(5,5))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.plot(x,y1, color='blue')
plt.plot(x,y2, color='blue')
plt.xlabel('x')
plt.ylabel('y')
```

#### Задание 3

Напишите код, который будет переводить полярные координаты в декартовы.


```python
def polar_to_decart(R, alpha):
    ''' Переводит полярные координаты (R - радиус, alpha - угол) в декартовые x и y
    '''
    x = R * np.cos(alpha)
    y = R * np.sin(alpha)
    return (x, y)
```

```python
# проверка
R = 3
alpha = np.pi / 3
x, y = polar_to_decart(R, alpha)
print(x, y)
```

#### Задание 4
Решите систему уравнений:  
$y = x^2 – 1$  
$e^x + x(1 – y) = 1$

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (-1, 0))
x2, y2 = fsolve(equations, (2, 5))
x3, y3 = fsolve(equations, (4, 15))

print(x1, y1)
print(x2, y2)
print(x3, y3)

x = np.linspace(-3, 5, 401)
plt.plot(x, (x**2-1), label='y=x^2-1')
plt.plot(x, ((exp(x) + x - 1)/x), label='y=(e^x+x-1)/x')
plt.grid()
plt.plot(x1, y1, 'o', color='red')
plt.plot(x2, y2, 'o', color='red')
plt.plot(x3, y3, 'o', color='red')
plt.legend();
```
