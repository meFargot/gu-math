Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20). Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
def vec_len(vec):
    """ Расчитывает длину вектора через цикл
    """
    dims = len(vec)
    sum_sq = 0
    for i in range(dims):
        sum_sq += vec[i] ** 2
    return np.sqrt(sum_sq)
```

```python
def vec_len_2(vec):
    """ Расчитывает длину вектора без цикла
    """
    return np.sqrt(np.sum([coord ** 2 for coord in vec]))
```

##### Проверка

```python
# зададим вектора
x1 = np.array([20,20,20])
x2 = np.array([0,0,-20])
```

```python
# расчет по формуле
np.sqrt(x1[0]**2 + x1[1]**2 + x1[2]**2)
```

```python
# расчет с помощью написанных функций
vec_len(x1)
```

```python
vec_len_2(x1)
```

#### Задание 3
Напишите код на Python, реализующий построение графиков:
1) окружности,
2) эллипса,
3) гиперболы.

##### Окружность

```python
x = []
y1 = []
y2 = []
R = 3
for i in range(-5000, 5001):
    x_ = i/1000
    x.append(x_)
    y1.append(np.sqrt(R**2 - x_**2))
    y2.append(-np.sqrt(R**2 - x_**2))
plt.figure(figsize=(5,5))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.plot(x,y1, color='blue')
plt.plot(x,y2, color='blue')
plt.xlabel('x')
plt.ylabel('y')
```

##### Эллипс

```python
x = []
y1 = []
y2 = []
a = 2
b = 1
for i in range(-5000, 5001):
    x_ = i/1000
    x.append(x_)
    y1.append(np.sqrt( b*(1 - (x_**2/a**2)) ))
    y2.append(-np.sqrt( b*(1 - (x_**2/a**2)) ))
plt.figure(figsize=(5,5))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.plot(x,y1, color='blue')
plt.plot(x,y2, color='blue')
plt.xlabel('x')
plt.ylabel('y')
```

##### Гипербола

```python
x = []
y1 = []
y2 = []
a = 1
b = 1
for i in range(-5000, 5001):
    x_ = i/1000
    x.append(x_)
    y1.append(np.sqrt( b*((x_**2/a**2) - 1) ))
    y2.append(-np.sqrt( b*((x_**2/a**2) - 1) ))
plt.figure(figsize=(5,5))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.plot(x,y1, color='blue')
plt.plot(x,y2, color='blue')
plt.xlabel('x')
plt.ylabel('y')
```

#### Задание 5
Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)

X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X1, Y1 = np.meshgrid(X1, Y1)
Z1 = 2*X1 + 3*Y1

X2 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-5, 5, 0.5)
X2, Y2 = np.meshgrid(X2, Y2)
Z2 = 2*X2 + 3*Y2 + 10

ax.plot_surface(X1, Y1, Z1)
ax.plot_surface(X2, Y2, Z2)
show()
```

Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)

X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X1, Y1 = np.meshgrid(X1, Y1)
Z1 = X1**2/2**2 + Y1**2/3**2  # эллиптический параболоид

X2 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-5, 5, 0.5)
X2, Y2 = np.meshgrid(X2, Y2)
Z2 = X2**2/3**2 - Y2**2/2**2  # гиперболический параболоид

ax.plot_surface(X1, Y1, Z1)
ax.plot_surface(X2, Y2, Z2)

show()
```
