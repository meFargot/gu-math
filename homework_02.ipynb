#### Задание 3

Напишите код, который будет переводить полярные координаты в декартовы.


```python
def polar_to_decart(R, alpha):
    ''' Переводит полярные координаты (R - радиус, alpha - угол) в декартовые x и y
    '''
    x = R * np.cos(alpha)
    y = R * np.sin(alpha)
    return (x, y)
```

```python
# проверка
R = 3
alpha = np.pi / 3
x, y = polar_to_decart(R, alpha)
print(x, y)
```

#### Задание 4
Решите систему уравнений:  
$y = x^2 – 1$  
$e^x + x(1 – y) = 1$

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (-1, 0))
x2, y2 = fsolve(equations, (2, 5))
x3, y3 = fsolve(equations, (4, 15))

print(x1, y1)
print(x2, y2)
print(x3, y3)

x = np.linspace(-3, 5, 401)
plt.plot(x, (x**2-1), label='y=x^2-1')
plt.plot(x, ((exp(x) + x - 1)/x), label='y=(e^x+x-1)/x')
plt.grid()
plt.plot(x1, y1, 'o', color='red')
plt.plot(x2, y2, 'o', color='red')
plt.plot(x3, y3, 'o', color='red')
plt.legend();
```

#### Задание 3

1) Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) и сравните результаты.

```python
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 2:
        k = k + 1
monte_karlo_1 = k/n
print(k, n, k/n)
```

```python
from scipy.special import factorial as fl


def C_n_k(n, k):
    return fl(n) / (fl(k) * fl(n - k))


def bernulli(n, k, p):
    q = 1 - p
    return C_n_k(n, k) * p**k * q**(n-k)
```

```python
bern_1 = bernulli(4, 2, 0.5)
print(bern_1)
```

```python
abs(bern_1 - monte_karlo_1)
```
